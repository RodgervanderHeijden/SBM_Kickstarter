```python
import pandas as pd
import re
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from langdetect import detect

import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
tech = pd.read_csv('sbm - added categories.csv', encoding='latin-1')
```

```python
tech.info()
```

```python
texts = tech['Project_description']
```

```python
texts[0]
```

**Main cleaning**

```python
clean_texts = []
len_texts = []
lemmatizer = WordNetLemmatizer()
count = 0
for text in texts:
    print(count)
    count += 1
    #string
    text = str(text)
    #remove wrong decoded '
    text = re.sub('<[^>]+>', '', text)
    #remove white spaces
    text = text.split()
    #remove emails and URLs
    text = [word for word in text if '@' not in word and 'http' not in word and 'www' not in word]
    text = ' '.join(text)
    #punctuation and lowercase
    text = text.strip().lower()
    #throw away all non-letters
    text = re.sub('[^a-zA-Z ]+', '', text)
    #check if English
    if detect(text) == 'en':
        #tokenize
        tokens = word_tokenize(text)
        #remove stopwords
        text = [i for i in tokens if not i in ENGLISH_STOP_WORDS]
        #lemmatize
        clean_text = []
        for word in text:
            clean_text.append(lemmatizer.lemmatize(word))
        clean_texts.append(clean_text)
        len_texts.append(len(clean_text))
    else: 
        print(text)
        clean_texts.append('')


    
```

```python
clean_texts = [texts if len(texts) > 50 else '' for texts in clean_texts ]
```

```python
len(clean_texts)
```

**Sparse term removal**

```python
all_words = [word for text in clean_texts for word in text]
len(set(all_words))
```

```python
word_counts = {}
for word in set(all_words):
    word_counts[word] = all_words.count(word)
```

```python
sparse_terms = []
for text in clean_texts:
    count = 0
    for word in text:
        if word_counts[word] < 5: #sparsity level of words in ALL texts
            text.remove(word)
            count += 1
    sparse_terms.append(count)
```

```python
final = []
for text in clean_texts:
        final.append(' '.join(text))
```

**Add the extra columns.**

```python
duration = []
for i in range(460):
    duration.append(int((tech['Deadline'][i]-tech['Launched_at'][i])/86400)) #/86400 to get to days
```

```python
update_frequency = []
for i in range(460):
    try:
        update_frequency.append(float(tech['Project_nb_updates'][i])/duration[i])
    except:
        update_frequency.append(0.0)
```

```python
tech['clean'] = final
tech['sparse_terms'] = sparse_terms
tech['duration'] = duration
tech['update_freq'] = update_frequency
```

**Remove all rows that do not have a description after cleaning**

```python
tech = tech[tech['clean'] != '']
```
