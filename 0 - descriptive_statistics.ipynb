**Descriptive statistics of the complete info datasets**

11 weeks of data on all categories.

```python
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
from IPython.display import Markdown, display
def printmd(string):
    display(Markdown(string))
```

```python
info21 = pd.read_csv('Data/CSV/info_week21.csv', encoding='latin-1')
info22 = pd.read_csv('Data/CSV/info_week22.csv', encoding='latin-1')
info23 = pd.read_csv('Data/CSV/info_week23.csv', encoding='latin-1') # => geeft een foutmelding want iets zit niet goed encoded
info24 = pd.read_csv('Data/CSV/info_week24.csv', encoding='latin-1')
info25 = pd.read_csv('Data/CSV/info_week25.csv', encoding='latin-1')
info26 = pd.read_csv('Data/CSV/info_week26.csv', encoding='latin-1')
info27 = pd.read_csv('Data/CSV/info_week27.csv', encoding='latin-1')
info28 = pd.read_csv('Data/CSV/info_week28.csv', encoding='latin-1')
info29 = pd.read_csv('Data/CSV/info_week29.csv', encoding='latin-1')
info30 = pd.read_csv('Data/CSV/info_week30.csv', encoding='latin-1')
info31 = pd.read_csv('Data/CSV/info_week31.csv', encoding='latin-1')
info32 = pd.read_csv('Data/CSV/info_week32.csv', encoding='latin-1')
```

```python
all_info = pd.concat([info21, info22, info24, info25, info26, info27, info28, info29, info30, info31, info32])
```

```python
columns = list(all_info.columns)

for column in columns:
    printmd('**{}**'.format(column))
    print(all_info[column].describe(), '\n\n')
```
