```python
import pandas as pd
```

**Find all technology rows**

```python
# Save all technology rows from each week to a csv file
for week in range(21,33):
    info = pd.read_csv('Data/CSV/info_week{}.csv'.format(week), encoding='latin-1')
    tech = info[info['Category'] == 'Technology']
    tech.to_csv('Tech/tech{}.csv'.format(week))
```

**Load all technology rows**

```python
# Load the technology files
tech21 = pd.read_csv('Tech/tech21.csv', encoding='latin-1')
tech22 = pd.read_csv('Tech/tech22.csv', encoding='latin-1')
tech23 = pd.read_csv('Tech/tech23.csv', encoding='latin-1')
tech24 = pd.read_csv('Tech/tech24.csv', encoding='latin-1')
tech25 = pd.read_csv('Tech/tech25.csv', encoding='latin-1')
tech26 = pd.read_csv('Tech/tech26.csv', encoding='latin-1')
tech27 = pd.read_csv('Tech/tech27.csv', encoding='latin-1')
tech28 = pd.read_csv('Tech/tech28.csv', encoding='latin-1')
tech29 = pd.read_csv('Tech/tech29.csv', encoding='latin-1')
tech30 = pd.read_csv('Tech/tech30.csv', encoding='latin-1')
tech31 = pd.read_csv('Tech/tech31.csv', encoding='latin-1')
tech32 = pd.read_csv('Tech/tech32.csv', encoding='latin-1')
```

```python
# Make one big dataframe of all technology files
technology = pd.concat([tech21, tech22, tech23, tech24, tech25, tech26, tech27, tech28, tech29, tech30, tech31, tech32], axis=0)
technology = technology.drop('Unnamed: 0.1', axis=1)
```

**Find the last time instance for each project**

Projects can be connected over weeks by using the project slug.

```python
# Create an index column
index = []
for i in range(4184):
    index.append(i)

tech['index'] = index
```

```python
# Find the last appearence for each project
indices = []
for project in range(len(tech)): 
    print(project)
    original_slug = tech['project_slug'][project]
    current_last_appearance = tech['index'][project]
    for compare_project in range(len(tech)):
        if tech['project_slug'][compare_project] == original_slug and tech['index'][compare_project] > current_last_appearance:
            indices.append(current_last_appearance)
            current_last_appearence = tech['index'][compare_project]
```

**Find indices to keep**

```python
to_keep = set(index)-set(indices)
len(to_keep)
```

```python
keep_or_not = [1 if number in to_keep else 0 for number in index]
```

```python
tech['keep'] = keep_or_not
```

**Create final dataframe**

```python
# Dataframe of rows that we should keep
final_tech = tech[tech['keep'] == 1]
```

```python
# Drop the columns that we created
final_tech = final_tech.drop(['index', 'keep'], axis = 1)
```

```python
final_tech.to_csv('final_tech.csv', index=False)
```
