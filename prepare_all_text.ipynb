```python
import pandas as pd
import re
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from langdetect import detect

import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
tech = pd.read_csv('final_tech.csv', encoding='latin-1')
```

```python
texts = tech['Project_description']
```

```python
practice_texts = ['hoi ik ben ellen mijn Ã©mail is ellen@hoi.nl en mijn website is https://lalala', 'hoi dit is rodger en zijn email is rodger@lala.nl']
```

```python
clean_texts = []
len_texts = []
lemmatizer = WordNetLemmatizer()
count = 0
for text in texts:
    print(count)
    count += 1
    #string
    text = str(text)
    #remove white spaces
    text = text.split()
    #remove emails and URLs
    text = [word for word in text if '@' not in word and 'http' not in word and 'www']
    text = ' '.join(text)
    #punctuation and lowercase
    text = text.strip().lower()
    #throw away all non-letters
    text = re.sub('[^a-zA-Z ]+', '', text)
    #check if English
    if detect(text) == 'en':
        #tokenize
        tokens = word_tokenize(text)
        #remove stopwords
        text = ([i for i in tokens if not i in ENGLISH_STOP_WORDS])
        #lemmatize
        clean_text = []
        for word in text:
            clean_text.append(lemmatizer.lemmatize(word))
        clean_texts.append(clean_text)
        len_texts.append(len(clean_text))
    else: print(text)


    
```
